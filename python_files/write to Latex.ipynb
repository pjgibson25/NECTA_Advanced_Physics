```python
badstuff = ['(i) ',
'(ii) ',
'(iii) ',
'(iv) ',
'(v) ',
'(vi) ']

intro_file = open('../Final_Project/intro1.txt','r')
intro = intro_file.read()
intro_file.close()

f = open('../Final_Project/all_qustions.tex','w')

f.write(intro)

##################################################################






##################################################################
last_topic = 'poop'
last_subtopic = 'poop'
n = 0


for index,row in df.iterrows():
    cur_topic = row['topic1']
    cur_subtopic = row['subtopic1']

    
    if last_subtopic != cur_subtopic:
        if n != 0:
            f.write('\\end{itemize}\n\n')
    
    if cur_topic != last_topic:
        f.write('\n\\section{'+str(cur_topic)+'}\n\n')

    if last_subtopic != cur_subtopic:
        f.write('\\subsection{'+str(cur_subtopic)+'}\n')
        f.write('\\begin{itemize}\n')
    
    f.write('\\item ')
    f.write('('+str(row['year'])+')  ')
    
    # Question stuff
    question_v1 = row['question']
    splitup = question_v1.split('\n')
    
    if (len(splitup)>1):
        if (len(splitup[1])>4):
            f.write(str(splitup[0]))
            f.write('\\begin{itemize}\n')
            for j in np.arange(1,len(splitup)):
                subquestion = str(splitup[j])
                for k in np.arange(0,len(badstuff)):
                    subquestion = subquestion.replace(badstuff[k],'')
                if len(subquestion)>5:
                    f.write('\\item ')
                    f.write(subquestion+'\n')
            f.write('\\end{itemize}\n')
    else:
        f.write(str(splitup[0])+'\n')
    #############################################################
    
    last_topic = cur_topic
    last_subtopic = cur_subtopic
    n+=1

    
f.write('\n\\end{itemize}\n\n')
f.write('\\end{document}')

f.close()
```
