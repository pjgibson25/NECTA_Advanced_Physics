```python
#################################################################

topics = np.array(['Measurement',
                  'Mechanics',
                  'Fluid Dynamics',
                  'Properties of Matter',
                  'Heat',
                  'Vibrations and Waves',
                  'Electrostatics',
                  'Electromagnetism',
                  'Current Electricity',
                  'Electronics',
                  'Atomic Physics',
                  'Environmental Physics'])

#################################################################
os.chdir(working_dir)
codee = open('subtopics_and_codes.txt','r')
codey = codee.read()
codee.close()
splitup = codey.split('\n')[:-1]
subtopics = []
subtopic_que = []
for i in np.arange(0,len(splitup)):
    if ((i+2) % 2 == 0):
        subtopics.append(splitup[i])
    else:
        subtopic_que.append(splitup[i])
        
subtopic_que = np.array(subtopic_que)
subtopics = np.array(subtopics)

pair = np.column_stack((subtopic_que,subtopics))

#################################################################

titles = np.chararray((len(pair),4), itemsize = 100, unicode=True)


prev_num = 0
point = 1

for i in np.arange(0,len(titles)):
    titles[i,0] = pair[i,0]
    titles[i,2] = pair[i,1]
    num = int(float(titles[i,0][:2]))
    
    
    titles[i,1] = topics[num-1]
    
    if num == prev_num:
        titles[i,3] = str(num)+'.'+str(point)
        point += 1
    else:
        point = 1
        titles[i,3] = str(num)+'.'+str(point)
        point +=1
    
    prev_num = num
    
#################################################################
```

```python
# Create a bunch of unique codes

stuffz = string.digits+string.ascii_uppercase
codes = []
for i in np.arange(0,len(stuffz)):
    for j in np.arange(0,len(stuffz)):
        for k in np.arange(0,len(stuffz)):
            codes.append((stuffz[i])+(stuffz[j])+(stuffz[k]))
            
```

```python
# Get all physics paper filenames

os.chdir(working_dir)
os.chdir('../LaTex_finaldraft/')
finished_txts = os.listdir()



# Find the number of questions, m
m = 0

for i in np.arange(0,len(finished_txts)):
    file = open(finished_txts[i],'r')
    text = file.read()
    file.close()
    questions = text.split('\n\n')[1:]
    for j in np.arange(0,len(questions)):
        m+=1

empty = np.chararray((m,12),itemsize=1000,unicode=True)
empty[:,:] = '.'
   
    
    
goofcount = 0
q = 0
for i in np.arange(0,len(finished_txts)):
    # open each test file and readit
    file = open(finished_txts[i],'r')
    text = file.read()
    file.close()
    
    # take important info from file name and split questions by \n\n indicator
    questions = text.split('\n\n')[1:]
    year = finished_txts[i][:4]
    version = finished_txts[i][6:-4]
    
    # loop through each question 
    for j in np.arange(0,len(questions)):
        
        # take code created before
        code = codes[q]
        # fill table (called 'empty') with basic info about question
        empty[q,0] = code
        empty[q,1] = questions[j]
        empty[q,2] = year
        empty[q,3] = version
        
        # want question in all uppercase for non-case sensitive keyword matching
        quest_up = questions[j].upper()
        
        # start with 0 keyword matches and loop through keywords
        num_matches = 0
        
        # loop through subtopic keys
        for k in np.arange(0,len(keyz)):
            
            # first get keyz row cooresponding to only one subtopic
            keylist = keyz[k]
            
            # start with 0 subtopic keyword matches
            key_ind = 0
            
            for m in np.arange(0,len(keylist)):
                
                # make keyword all caps for non-case sensitive keyword matching
                test = keylist[m].upper()
                
                # see if its in the question!  If so, make note in key_ind
                if test in quest_up:
                    key_ind += 1
                    
            # if there were 1+ subtopic matches, use 'titles' table to collect subtopic info for 'empty' table
            if key_ind != 0:
                
                ################################################################
                # Some questions might be tagged by multiple keywords that
                # coorespond to multiple subtopics.  My table allows for up to
                # 3 unique subtopic tags.  This was an arbitrary number I chose.
                ################################################################
                
                # If this is the 1st subtopic match, put [topic,subtopic] in columns [4,5]
                if num_matches == 0:
                    empty[q,4] = titles[k,1]
                    empty[q,5] = titles[k,2]
                    num_matches += 1
                    
                # If this is the 2nd subtopic match, put [topic,subtopic] in columns [6,7]
                elif num_matches == 1:
                    empty[q,6] = titles[k,1]
                    empty[q,7] = titles[k,2]
                    num_matches += 1
                    
                # If this is the 3rd subtopic match, put [topic,subtopic] in columns [8,9]
                elif num_matches == 2:
                    empty[q,8] = titles[k,1]
                    empty[q,9] = titles[k,2]
                    num_matches += 1
               
                # If there are 4+ subtopic matches, print out a GOOF error.
                else:
                    print('goof')
                    print(finished_txts[i])
                    print(questions[j])
                    print('---------------__')
                    goofcount += 1
                    
        ################################################################
        # Questions with three hashes denote that I was unable to enter 
        # the question into a .csv table.  These difficult questions 
        # have a '1' in the 10 column.
        #
        # Mostly figures and tables and complex chemistry notations
        ################################################################
        
        if '###' in quest_up:
            empty[q,10] = '1'
        else:
            empty[q,10] = '0'
        
        ################################################################
        # Questions with no $ in them denote questions with no math mode
        # notations (Latex).  Denoted with '1' in column 11.
        #
        # Likely definition problems and physics word problems.
        ################################################################
        
        if '$' not in quest_up:
            empty[q,11] = '1'
        else:
            empty[q,11] = '0'
        q +=1
```
